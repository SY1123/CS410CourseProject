```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import word_tokenize
import nltk.data
nltk.download('vader_lexicon')
nltk.download('punkt')
```

```python
#import and clean data
df = pd.read_csv("songs.csv", encoding='unicode_escape',sep=",")
df = df.drop(['Unnamed: 31',
       'Unnamed: 32', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35',
       'Unnamed: 36', 'Unnamed: 37', 'Unnamed: 38', 'Unnamed: 39',
       'Unnamed: 40', 'Unnamed: 41', 'Unnamed: 42', 'Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50', 'Unnamed: 51',
       'Unnamed: 52', 'Unnamed: 53', 'Unnamed: 54', 'Unnamed: 55',
       'Unnamed: 56', 'Unnamed: 57', 'Unnamed: 58', 'Unnamed: 59',
       'Unnamed: 60'],axis=1)
df = df[df['lyrics'].notna()]
df_ = df[df['lyrics'] != 'Error: Could not find lyrics.']
df_['lyrics'] = df_['lyrics'].str.replace("\n\n", ". ")
df_['lyrics'] = df_['lyrics'].str.replace("\n", ". ")
df_['lyrics'] = df_['lyrics'].str.lower().str.strip()
# ll = df_.sample(n = 10)['lyrics'].to_list()
# ll[8]
df_ = df_.sort_values(by='year', ascending=False)
df = df_
df = df.reset_index(drop=True)
df
```

```python
'''
The compound score is computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, 
and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). 
This is the most useful metric if you want a single unidimensional measure of sentiment for a given sentence. 
Calling it a 'normalized, weighted composite score' is accurate.

It is also useful for researchers who would like to set standardized thresholds for classifying sentences as either 
positive, neutral, or negative. Typical threshold values (used in the literature cited on this page) are:

positive sentiment: compound score >= 0.05
neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
negative sentiment: compound score <= -0.05

'''

negative = []
neutral = []
positive = []
compound = []
#Initialize the model
sid = SentimentIntensityAnalyzer()
#Iterate for each row of lyrics and append the scores
for i in df.index:
    scores = sid.polarity_scores(df['lyrics'].iloc[i])
    negative.append(scores['neg'])
    neutral.append(scores['neu'])
    positive.append(scores['pos'])
    compound.append(scores['compound'])
df['negative'] = negative
df['neutral'] = neutral
df['positive'] = positive
df['compound'] = compound
df.sample(n=10)

```

```python
sentiment_analyisis = []
for i in df.index:
    if df['compound'].iloc[i] < -0.7:
        sentiment_analyisis.append('negative+')
    elif df['compound'].iloc[i] <= -0.05:
        sentiment_analyisis.append('negative')
    elif df['compound'].iloc[i] > 0.7:
        sentiment_analyisis.append('positive+')
    elif df['compound'].iloc[i] >= 0.05:
        sentiment_analyisis.append('positive')
    else:
        sentiment_analyisis.append('neatral')
df['sentiment_analyisis'] = sentiment_analyisis
```
