```python
import pandas as pd
import math
```

```python
df = pd.read_csv('songs_sentiment.csv')
df = df[df['weeks'].notna()]
df = df[df['rank'].notna()]
df
```

```python
# I plan to use a Weighte Rating Formula to calculate weighted score 
# This score determine which songs should be recommended by default
# popularity = (avg ranking / ranking) * log(weeks + 1)

mean_rank = df['rank'].mean() #75.69595903681152

def weighted_rating(x):
    ranking = x['rank']
    weeks = x['weeks']
    # Calculation based on the IMDB formula
    return (mean_rank/ ranking) * math.log(weeks + 1)

df['score_recomm'] = df.apply(weighted_rating, axis=1)
df = df.sort_values('score_recomm', ascending=False)
# df = df.drop('Unnamed: 0', axis=1)  
df.to_csv('ccc.csv')
```
